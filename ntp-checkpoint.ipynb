```python
time=['Months','Minute','Hours','Year','Day']
sayi=['counts']
modul=['month','minute','hour','year','day']
def zaman (x,sayi,z,accident_df_2): 
    for i in x:
        yield i
    for j in z:
        yield j
       
    return accident_df_2.apply(accident_df_2[i] == pd.to_datetime(accident_df_2[sayi]).dt.j)    
uygula=map(zaman,(time),(modul))   
print(uygula)
```

```python
b = [0,4,8,12,16,20,24]
l = ['Late_Night', 'Early_Morning','Morning','Noon','Eve','Night']
accident_df_3['Araliklar'] = pd.cut(accident_df_3['Hours'].astype(int), bins=b, labels=l, include_lowest=True)
```

```python
def getDayNumber(year, month, day):
    weekdays = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
    dayNumber = calendar.weekday(year, month, day)
    return (weekdays[dayNumber])



accident_df_3['weekday']= accident_df_3.apply(lambda x: getDayNumber(x['Year'], x['Months'], x['Day']), axis=1)

accident_df_3
```
